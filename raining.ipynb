```python
import pandas as pd

# 指定Excel檔案路徑
excel_file = "raining.xlsx"

# 使用pandas讀取Excel檔案
df = pd.read_excel(excel_file)

# 輸出DataFrame的內容
print(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 修改欄位名稱和內容
df = df.rename(columns={'年份': 'Year', '平均降雨量(ml)': 'Average Rainfall'})

# 提取年份和平均降雨量資料
years = df['Year']
rainfall = df['Average Rainfall']

# 標示每個資料的數字
for i, j in zip(years, rainfall):
    plt.text(i, j, str(j), ha='center', va='bottom')

# 繪製圖表
plt.plot(years, rainfall, marker='o', color='#541675')
plt.xlabel('Year')
plt.ylabel('Average Rainfall')
plt.title('Average rainfall for 10 years in Taiwan')
plt.grid(True)
plt.show()
```
